```python
# Importing the necessary modules 
import os 
import cv2 
import random 
import joblib
import numpy as np 
import seaborn as sns
from tqdm import tqdm 
import tensorflow as tf 
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder  
from tensorflow.keras.utils import to_categorical 
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models, backend as K
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
```

<h2> Fingerprint Data Processing Module </h2>

```python
# Creating a class for the finger print data process 
class FingerprintDataProcessor: 
    # Initilizing the init method 
    def __init__(self, dataDir, imgSize=(160, 160)): 
        self.dataDir = dataDir
        self.imgSize = imgSize 
        self.labelEncoder = LabelEncoder() 

    # Load and preprocess the image 
    def loadAndPreprocessImage(self, imagePath): 
        # Load the image using opencv 
        img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE) 

        # Checking if the image is None type 
        if img is None: 
            return None 
            
        # Resize the image 
        img = cv2.resize(img, self.imSize)
        img = img.astype("float32") / 255.0 
        img = np.expand_dims(img, axis=-1) 

        # Return the image 
        return img 

    # Load the dataset 
    def loadDataset(self): 
        # Load all the images and labels 
        images = []
        labels = []

        # for values in the data directory 
        for labelName in os.listdir(self.dataDir): 
            # Skipping the ".DS_Store" file for macOS systems
            if (labelName == ".DS_Store") or (labelName.startswith(".")):
                continue
                
            # Get the full path to the label directory 
            labelDir = os.path.join(self.dataDir, labelName)

            # if the label directory is not a folder, continue 
            if not os.path.isdir(labelDir): 
                continue 

            # Displaying the status 
            print(f"Loading {labelName} fingerprints...") 

            # for the image files in the label directory 
            for imageFile in tqdm(os.listdir(labelDir)): 
                # Ensuring the images end with ".png, .jpg, .jpeg, and .bmp" 
                if imageFile.lower().endswith((".png", ".jpg", ".jpeg", ".bmp")): 
                    # Getting the image path 
                    imgPath = os.path.join(labelDir, imgFile)

                    # Load and preprocess the image 
                    img = self.loadAndPreprocessImage(imgPath) 

                    # if the image is not None, execute the block of code below 
                    if img is not None: 
                        images.append(img)
                        labels.append(labelName) 

        # Returning the numpy array for the labels and images 
        return np.array(images), np.array(labels) 

    # Creating a function for the triplets 
    def createTriplets(self, images, labels): 
        # Create triplets (anchor, positive, negative) for Siamese network 
        triplets = [] 
        labelToIndices = {} 

        # Group indices by label 
        for idx, label in enumerate(labels): 
            # if the label does not exist in the labelToIndices 
            if label not in labelToIndices: 
                labelToIndices[label] = []
            labelToIndices[label].append(idx) 

        # Create triplets 
        for anchorIdx, anchorLabel in enumerate(labels): 
            # Get positive indices (same label, different image) 
            positiveIndices = [idx for idx in labelToIndices[anchorLabel] if idx != anchorIdx]

            # if the length of positve indices is equal to zero 
            if len(positiveIndices) == 0: 
                continue 

            # Selecting a random value using the random.choice inbuilt method 
            positiveIdx = np.random.choice(positiveIndices) 

            # Get negative indices (different label) 
            negativeLabels = [l for l in labelToIndices.keys() if l != anchorLabel] 

            # If the length for the negative labels is equal to zero
            if len(negativeLabels) == 0: 
                continue 

            # Getting the negative label, and the negative idx 
            negativeLabel = np.random.choice(negativeLabels) 
            negativeIdx = np.random.choice(labelToIndices[negativeLabel]) 

            # Appending the values into the triplets array 
            triplets.append((anchorIdx, positiveIdx, negativeIdx)) 

        # Returning the triplets array 
        return np.array(triplets) 

    # Creating a method for generating the batches 
    def generateBatches(self, images, triplets, batchSize=32): 
        # Generate the batches of triplets for training 
        numTriplets = len(triplets) 
        indices = np.random.permutation(numTriplets) 

        # For the range of values starting from zero through the num triplets 
        # to the batch size 
        for start in range(0, numTriplets, batchSize): 
            end = min(start + batchSize, numTriplets) 
            batchIndices = indices[start:end]

            # Setting the batch anchors, batch positives 
            # and batch negatives 
            batchAnchors = [] 
            batchPositives = [] 
            batchNegatives = [] 

            # For idx in batch indices 
            for idx in batchIndices: 
                # Get the anchor idx, positive idx, and negative idx values 
                anchorIdx, positiveIdx, negativeIdx = triplets[idx]
                batchAnchors.append(images[anchorIdx]) 
                batchPositives.append(images[positiveIdx]) 
                batchNegatives.append(images[negativeIdx]) 

            # Yield the values as a numpy array 
            yield np.array(batchAnchors), np.array(batchPositives), np.array(batchNegatives)                      
```

<h2> Siamese Network Model </h2>

```python
# Creating the SiameseNetwork 
class SiameseNetwork: 
    # Initializing the init method 
    def __init__(self, inputShape=(160, 160, 1), embeddingDim=128): 
        # Setting the input shape 
        self.inputShape = inputShape 
        self.embeddingDim = embedingDim 
        self.embeddingModel = self.buildEmbeddingModel() 

    # Creating the build embedding model method 
    def buildEmbeddingModel(self): 
        # Build the embedding model using MobileNetV2 as a backbone 
        inputLayer = layers.Input(shape=self.inputShape) 

        # Convert grayscale to 3 channels for pre-trained model 
        if self.inputShape[-1] == 1:
            x = layers.Conv2D(3, (1, 1), padding="same")(inputLayer) 

        # Else if the input shape is not equals to 1 
        else: 
            # Set the value of x as the input layers 
            x = inputLayer 

        # Use MobileNetV2 as a feature extractor 
        baseModel = MobileNetV2(
            inputShape=(self.inputShape[0], self.inputShape[1], 3), 
            include_top=False, 
            weights="imagenet"
        )

        # Setting the base model trainable as false 
        baseModel.trainable = False 

        # Building the neural network 
        x = baseModel(x) 
        x = layers.GlobalAveragePooling2D()(x) 
        x = layers.Dense(512, activation="relu")(x) 
        x = layers.BatchNormalization()(x) 
        x = layers.Dropout(0.3)(x) 
        x = layers.Dense(256, activation="relu")(x) 
        x = layers.BatchNormalization()(x) 

        # Setting the embeddings 
        embeddings = layers.Dense(self.embeddingDim, name="embeddings")(x) 

        # Building the model 
        model = models.Model(inputs=inputLayer, outputs=embeddings) 

        # Returing the model 
        return model 

    # Building the triplet loss method 
    def tripletLoss(self, margin=1.0): 
        # Triplet loss function for Siamese network 
        def loss(yTrue, yPred): 
            # ypred contains anchor, positive, negative embeddings concatenated 
            anchor, positive, negative = tf.split(yPred, 3, axis=1) 

            # Calculate distances 
            posDistance = tf.reduce_sum(tf.square(anchor - positive), axis=1) 
            negDistance = tf.reduce_sum(tf.square(anchor - negative), axis=1)

            # Triplet loss 
            basicLoss = posDistance - negDistance + margin
            loss = tf.reduce_mean(tf.maximum(basicLoss, 0.0)) 

            # Return the loss 
            return loss; 

        # Returing the loss 
        return loss; 

    # Building the siamese model method 
    def buildSiameseModel(self): 
        # Build the complete siamese model 
        # Input layers for triplets 
        anchorInput = layers.Input(shape=self.inputShape, name="anchor") 
        positiveInput = layers.Input(shape=self.inputShape, name="positive") 
        negativeInput = layers.Input(shape=self.inputShape, name="negative") 

        # Get embeddings 
        anchorEmbeddings = self.embeddingModel(anchorInput) 
        positiveEmbedding = self.embeddingModel(positiveInput) 
        negativeEmbedding = self.embeddingModel(negativeInput) 

        # Concatenate embeddings 
        concatenated = layers.Concatenate(axis=1)([
            anchorEmbeddings, positiveEmbedding, negativeEmbedding
        ])

        # Building the model 
        model = models.Model(
            inputs=[anchorInput, positiveInput, negativeInput], 
            outputs=concatenated
        )

        # Return the model 
        return model 

    # Compiling the model 
    def compileModel(self, learningRate=0.0001): 
        # Compile the siamese model 
        model = self.buildSiameseModel() 
        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=learningRate), 
            loss=self.tripletLoss() 
        )

        # Return the model 
        return model 
```

<h2> Fingerprint Recognizer </h2>

```python
# Creating a class for the finger print recognizer 
class FingerprintRecognizer: 
    # Creating the init method 
    def __init__(self, embeddingModel, labelEncoder, threshold=0.5): 
        # Getting the respective values 
        self.embeddingModel = embeddingModel
        self.labelEncoder = labelEncoder 
        self.threshold = threshold 
        self.referenceEmbeddings = {} 
        self.referenceLabels = [] 

    # Generate the embedding
    def generateEmbedding(self, image):
        # Generate embedding for a single image 
        if len(image.shape) == 3: 
            image = np.expand_dims(image, axis=0) 

        # Get the embeddings 
        embedding = self.embeddingModel.predict(image, verbose=0) 

        # Returning the embeding 
        return embedding[0]

    # Creating the register fingerprint method 
    def registerFingerPrint(self, image, label): 
        # Register a fingerprint with its label 
        embedding = self.generateEmbedding(image) 

        # if the label is not in the reference embeddings 
        if label not in self.referenceEmbeddings: 
            self.referenceEmbeddings[label] = []

        # Append the embedding 
        self.referenceEmbeddings[label].append(embedding) 

    # Creating a method for computing the similarity 
    def computeSimilarity(self, embedding1, embedding2): 
        # Compute cosine similarity between two embeddings 
        embedding1 = embedding1 / np.linalg.norm(embedding1) 
        embedding2 = embedding1 / np.linalg.norm(embedding2) 

        # Compute the similarity 
        similarity = np.dot(embedding1, embedding2) 

        # return the similarity 
        return similarity; 

    # Create a method for predicting 
    def predict(self, image): 
        # Predict the label for a given fingerprint image 
        queryEmbedding = self.generateEmbedding(image) 

        # Setting the best match as None type 
        bestMatch = None 
        bestSimilarity = -1 

        # Looping through the reference embeddings 
        for label, embeddings in self.referenceEmbeddings.items(): 
            # Compare with all registered embeddings of this label 
            similarities = [self.computeSimilarity(queryEmbedding, refEmb) 
                            for refEmb in embeddings]

            # Setting the max similarity 
            maxSimilarity = max(similarities) if similarities else -1 

            # If the max similarity is greater than the best similarity 
            if maxSimilarity > bestSimilarity: 
                bestSimilarity = maxSimilarity
                bestMatch = label 

        # if the bestSimilarity is greater than the threshold 
        if (bestSimilarity > self.threshold): 
            # Return both the best match, and the best similarity 
            return bestMatch, bestSimilarity 

        # Else, return "unknow", with the best similarity 
        else: 
            # Return the values 
            return "unknown", bestSimilarity; 
```

<h2> Training Pipeline </h2>

```python
# Creating the class for the training pipeline 
class TrainingPipeline: 
    # Creating the init method 
    def __init__(self, dataProcessor, siameseNetwork): 
        # Getting the values 
        self.dataProcessor = dataProcessor 
        self.siameseNetwork = siameseNetwork 

    # Creating the method for training the model 
    def train(self, epochs=50, batchSize=32, validationSplit=0.2): 
        # Train the siamese network 
        print("[INFO]: Loading the dataset...")
        images, labels = self.dataProcessor.loadDataset() 

        # Displaying the loaded images 
        print(f"[INFO]: Loaded {len(images)} images with {len(np.unique(labels))} classes") 

        # Split the dataset 
        trainIdx, valIdx = train_test_split(
            range(len(images)), 
            test_size=validationSplit, 
            stratify=labels, 
            random_state=42 
        )

        # Getting the training images, labels and validation values 
        trainImages = images[trainIdx] 
        trainLabels = labels[trainIdx] 
        valImages = images[valIdx] 
        valLabels = labels[valIdx] 

        # Displaying the length of the training samples and 
        # the validation samples 
        print(f"Training smaples: {len(trainImages)}")
        print(f"Validation samples: {len(valImages)}") 

        # Create triplets 
        print("[INFO]: Creating training triplets...") 
        trainTriplets = self.dataProcessor.createTriplets(trainImages, trainLabels) 

        # Displaying the length of the train triplets 
        print(f"[INFO]: Created {len(trainTriplets)} training triplets") 

        valTriplets = self.dataProcessor.createTriplets(valImages, valLabels) 

        # Build and compile model 
        model = self.siameseNetwork.compileModel()

        # Setting the callbacks 
        callbacks = [
            ModelCheckpoint(
                "models/bestSiameseModel.h5", 
                save_best_only=True, 
                monitor="val_loss", 
                mode="min"
            ), 
            EarlyStopping(
                monitor="val_loss", 
                patience=10, 
                restore_best_weights=True 
            ), 
            ReduceLROnPlateau(
                monitor="val_loss", 
                factor=0.5, 
                patience=5, 
                min_lr=1e-7
            )
        ]

        # Training the model 
        print("[INFO]: Starting the training session...") 
        history = model.fit(
            self.dataProcessor.generateBatches(trainImages, trainTriplets, batchSize),
            steps_per_epoch=len(trainTriplets), 
            validation_data=self.dataProcessor.generateBatches(valImages, valTriplets, batchSize), 
            validation_steps=len(valTriplets),
            epochs=epochs, 
            callbacks=callbacks, 
            verbose=1
        )

        # Return the model and history 
        return model, history 

    # Creating a method for saving the model 
    def saveModel(self, model, recognizer, modelPath="models/", recognizerPath="models/"): 
        # Save the trained model and recognizer 
        os.makedirs(modelPath, exist_ok=True) 
        os.makedirs(recognizerPath, exists_ok=True) 

        # Saving the embedding model 
        model.save(os.path.join(modePath, 'embeddingModel.h5'))

        # Save the recognizer 
        joblib.dump(recognizer, os.path.join(recognizerPath, 'fingerprintRecogizer.pkl')) 

        # Save the label encoder 
        joblib.dump(self.dataProcessor.labelEncoder, 
            os.path.join(recognizerPath, 'labelEncoder.pkl')) 

        # Displaying the status 
        print("[INFO]: Models saved successfully!")

    # Creating a method for registering the finger prints 
    def registerFingerprintsForRecognizer(self, recognizer, images, labels, samplesPerClass=10):
        # Register fingerprints for the recognizer 
        uniqueLabels = np.unique(labels) 

        # For labels in the unique labels 
        for label in uniqueLabels: 
            labelIndices = np.where(labels == label)[0]

            # Use first N samples for registration 
            registerIndices = labelIndices[:samplesPerClass] 

            # For idx in register indices 
            for idx in registerIndices:
                recognizer.registerFingerprintsForRecognizer(images[idx], label) 

            # Displaying the status 
            print(f"Registered {len(registerIndices)} fingerprints for {label}") 

        # Returning the recognizer 
        return recognizer           
```

```python
# Creating a function for preprocessing the image for inference 
def preprocessImageForInference(imagePath, imageSize=(160, 160)): 
    # Preprocess image for inference 
    img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE) 

    # If the img is None 
    if img is None: 
        print(f"[INFO]: Error loading the image: {imagPath}") 
        return None 

    # Resizing the image 
    img = cv2.resize(img, imgSize) 
    img = img.astype('float32') / 255.0 
    img = np.expand_dims(img, axis=-1) 

    # Returning the resized image 
    return img 

# Creating a function for visualizing the embeddings 
def visualizeEmbeddings(embeddings, labels, title="Fingerprint Embeddings Visualization"): 
    # Visualize embeddings using principal component analysis 
    pca = PCA(n_components=2) 
    embeddings2D = pca.fit_transform(embeddings) 

    # Setting the figure dimensions 
    plt.figure(figsize=(12, 8)) 
    uniqueLabels = np.unique(labels) 
    colors = plt.cm.rainbow(np.linspace(0, 1, len(uniqueLabels)))

    # For the labels and colors in the unique labels and colors, 
    # get the mask, and plot a scatter plot of the embeddings 
    for label, color in zip(uniqueLabels, colors): 
        mask = labels == label

        # Display the scatter plot 
        plt.scatter(embeddings2D[mask, 0], embeddings2D[mask, 1], 
                    c=[color], label=label, alpha=0.7)

    # Plotting the graph 
    plt.title(title) 
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2") 
    plt.legend() 
    plt.grid(True, alpha=.3)
    plt.show() 

# Creating a function for evaluating the model 
def evaluateModel(recognizer, testImages, testLabels, threshold=.5): 
    # Evaluate the fingerprint recognition system 
    predictions = [] 
    confidences = [] 

    # for img, and true label in testImage, and testlabels 
    for img, trueLabel in zip(testImages, testLabels): 
        predLabel, confidence = recognizer.predict(img) 
        predictions.append(predLabel)
        confidences.append(confidence) 

    # Generate metrics 
    print(f"Classification Report: ")
    print(classification_report(testLabels, predictions))

    # Confusion Matrix 
    cm = confusion_matrix(testLabels, predictions, 
                         labels=np.unique(np.concatenate([testLabels, predictions])))

    # Setting the dimensions of the plot 
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", 
                cmap="Blues", 
                xticklabels=np.unique(np.concatenate([testLabels, predictions])), 
                yticklabels=np.unique(np.concatenate([testLabels, predictions])))

    # Setting the x and y axis values 
    plt.title("Confusion Matrix") 
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label") 
    plt.show() 

    # Return the predictions 
    return predictions, confidences
```
